#### Plotting from Database 2

````python
from collections import Counter

plt.style.use("fivethirtyeight")
plt.figure(figsize=(9,5))

ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

# to reverse the bar chart
languages.reverse()     
popularity.reverse()

plt.barh(languages, popularity) # horizintal chart

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()
````

- ### Stack Plots (Area Plots)

````python

minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

Arima = [1, 2, 3, 3, 4, 4, 4, 4, 5]
Neural = [1, 1, 1, 1, 2, 2, 2, 3, 4]
Deep = [1, 1, 1, 2, 2, 2, 3, 3, 3]

labels = ['Arima', 'Neural', 'Deep']
colors = ['lightcoral','cornsilk','powderblue']

plt.figure(figsize=(9,5))
plt.stackplot(minutes, Arima, Neural, Deep, labels=labels, colors=colors)

plt.title("Baby Stack Plot")
plt.tight_layout()

plt.legend(loc='upper left')
plt.show()
````

````python
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ['player1', 'player2', 'player3']
colors = ['#FFE4C4', '#001B3B', '#76b5c5']

plt.figure(figsize=(8,5))
plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)

plt.legend(loc=(0.07, 0.05))

plt.title("My Awesome Stack Plot")
plt.tight_layout()
plt.show()
````

- ## Plotting Time Series Data

````python
url2= "https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/08-TimeSeries/data.csv"
data5 = pd.read_csv(url2)
data5.head()
````

````python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
plt.style.use('seaborn')

data5['Date'] = pd.to_datetime(data5['Date'])
data5.sort_values('Date', inplace=True)

price_date = data5['Date']
price_close = data5['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()    # the ticks with the points in autoformat
date_format = mpl_dates.DateFormatter('%b, %d %Y')

plt.gca().xaxis.set_major_formatter(date_format )

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
````
